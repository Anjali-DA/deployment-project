```python
new_df = df.drop_duplicates(inplace=False)
new_df.shape
```

```python
new_df.isnull().sum()
```

```python
new_df['hospital_associated_infections'].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr_matrix = new_df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
new_df['hospital_associated_infections']= new_df['hospital_associated_infections'].replace(['sepsis', 'yes','UTI', 'mrsa', 'c-diff'], 'True')
```

```python
new_df['hospital_associated_infections']= new_df['hospital_associated_infections'].replace(['no', 'none'], 'False')
```

```python
title_mapping = {"False": 0, "True": 1}
new_df['hospital_associated_infections'] = new_df['hospital_associated_infections'].map(title_mapping)
```

```python
new_df.head()
```

```python
new_df['urine_report'].unique()
```

```python
urine_mapping = {"abnormal": 0, "normal": 1}
new_df['urine_report'] = new_df['urine_report'].map(urine_mapping)
```

```python
blood_mapping = {"abnormal": 0, "normal": 1}
new_df['blood_report'] = new_df['blood_report'].map(blood_mapping)
```

```python
new_df.head()
```

```python
plt.figure(figsize=(16,10))
sns.lineplot(y='hospital_associated_infections',x="blood_report",data=new_df)
plt.title('HAI',fontsize=20)
plt.xlabel('infections associated with blood report',fontsize=20)
plt.ylabel('blood report',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.lineplot(y='blood_pressure',x="wbc_count",hue='hospital_associated_infections',data=new_df)
plt.title('infections associated with blood report',fontsize=20)
plt.xlabel('WBC count',fontsize=20)
plt.ylabel('blood pressure',fontsize=20)
plt.show()
```
